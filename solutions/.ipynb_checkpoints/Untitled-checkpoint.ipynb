```python
def sum_of_squares(_from, _to):
    return sum(i**2 for i in range(_from, _to +1))

def square_of_sum(_from, _to):
    return sum(range(_from, _to +1))**2
```

```python
_from = 1
_to = 100
result = square_of_sum(_from, _to) - sum_of_squares(_from, _to)
```

```python
def count_leap_years(_from, _to):
    return sum(True for i in range(_from, _to+1) if i%400==0 or (i%4==0 and i%100!=0))

def total_number_of_days(_from, _to):
    return 365*(_to-_from+1) + count_leap_years(_from, _to)
```

```python
_from = 1900
_to = 2000
```

```python
tot = total_number_of_days(1900,2000)
```

```python
def find_list_of_multiples(n, _from, _to):
    return [i for i in range(_from, _to+1) if i%n==0]
```

```python
multiples = find_list_of_multiples(7, 1, tot)
```

```python
dict_months = {
    "january": 31,
    "february": 28,
    "march": 31,
    "april": 30,
    "may": 31,
    "june": 30,
    "july": 31,
    "august": 31,
    "september": 31,
    "october": 31,
    "november": 30,
    "december": 31
}

def find_list_first_day_month(_from, _to):
    lst = []
    sum = 1
    for i in range(_from, _to+1):
        for k in dict_months:
            if (i % 400 == 0 or (i %4 == 0 and i % 100 != 0 ))  and k == "february":
                sum += dict_months[k] + 1
            else: 
                sum += dict_months[k]
            lst.append(sum)
    return lst
```

```python
lst_2 = find_list_first_day_month(_from, _to)
```

```python
len(set(multiples).intersection(set(lst_2)))
```
